```python
selected_columns = ['REF_AREA_LABEL',
                   'TIME_PERIOD',
                   'INDICATOR_LABEL',
                   'OBS_VALUE'
                   ]
df_simple = df[selected_columns]
df_simple.head()
```

```python
df_simple.info()
```

```python
df_simple.isnull().sum()
```

```python
df_clean = df_simple.dropna()
```

```python
df_clean.info()
```

```python
df_clean['INDICATOR_LABEL'].value_counts()
```

```python
indicator_name = "GEM Adult Population Survey: Perceived opportunities"
df_indicator = df_clean[df_clean['INDICATOR_LABEL'] == indicator_name]
```

```python
df_indicator.info()
```

```python
df_indicator['TIME_PERIOD'].min(),df_indicator['TIME_PERIOD'].max()
```

```python
df_indicator['REF_AREA_LABEL'].nunique()
```

```python
df_indicator['OBS_VALUE'].describe()
```

```python
avg_by_year = df_indicator.groupby('TIME_PERIOD')['OBS_VALUE'].mean()
avg_by_year.head()
```

```python
avg_by_year.plot()
```

```python
before_2020 = df_indicator[df_indicator['TIME_PERIOD'] < 2020]['OBS_VALUE'].mean()
after_2020 = df_indicator[df_indicator['TIME_PERIOD'] >= 2020]['OBS_VALUE'].mean()
before_2020, after_2020
```

```python
country_before = (
    df_indicator[df_indicator['TIME_PERIOD'] < 2020]
    .groupby('REF_AREA_LABEL')['OBS_VALUE']
    .mean()
)

country_after = (
    df_indicator[df_indicator['TIME_PERIOD'] >= 2020]
    .groupby('REF_AREA_LABEL')['OBS_VALUE']
    .mean()
)
```

```python
comparison = (country_after - country_before).dropna()
comparison.sort_values(ascending=False).head(10)
```

```python
top5 = comparison.sort_values(ascending=False).head(5)
top5
```

```python
top5.plot(kind='bar',title ='Top 5 Countries by Increase in Perceived Opportunies')
```
